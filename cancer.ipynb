```python
def models(x_train,y_train):
    
    from sklearn.linear_model import LogisticRegression
    logist=LogisticRegression()
    logist.fit(x_train,y_train)
    
    from sklearn.neighbors import KNeighborsClassifier
    knn=KNeighborsClassifier(n_neighbors=5, metric='minkowski',p=2)
    knn.fit(x_train,y_train)
    
    from sklearn.ensemble import RandomForestClassifier  
    RandFor= RandomForestClassifier(n_estimators= 10, criterion="entropy")  
    RandFor.fit(x_train, y_train) 
    
    from sklearn.naive_bayes import GaussianNB
    NB=GaussianNB()
    NB.fit(x_train,y_train)
    
    from sklearn.svm import SVC
    svc_lin=SVC(kernel='linear',random_state=0)
    svc_lin.fit(x_train,y_train)
 
    from sklearn.svm import SVC
    svc_rbf=SVC(kernel='rbf',random_state=0)
    svc_rbf.fit(x_train,y_train)

    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=0)
    tree.fit(x_train,y_train)
    
    import xgboost as xgb
    xgboost = xgb.XGBClassifier(base_score=0.5, booster='gbtree')
    xgboost.fit(x_train, y_train)

    print("[0] Accuracy Score of Logistic Regression Accuracy: ",logist.score(x_train,y_train))
    print("[1] Accuracy Score of KNN Accuracy: ",knn.score(x_train,y_train))
    print("[2] Accuracy Score of Random Forest Accuracy: ",RandFor.score(x_train,y_train))
    print("[3] Accuracy Score of SVC- Linear Accuracy: ",svc_lin.score(x_train,y_train))
    print("[4] Accuracy Score of SVC- RBF Accuracy: ",svc_rbf.score(x_train,y_train))
    print("[5] Accuracy Score of Naviie Bayes Accuracy: ",NB.score(x_train,y_train))
    print("[6] Accuracy Score of Dicision Tree Accuracy: ",tree.score(x_train,y_train))
    print("[7] Accuracy Score of XG Boost Accuracy: ",xgboost.score(x_train,y_train))
    
    return logist,knn,RandFor,svc_lin,svc_rbf,NB,tree,xgboost
    
```

```python
model=models(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

for i in range(len(model)):
    cm=confusion_matrix(y_test,model[i].predict(x_test))
    
    print('Model[{}]'.format(i,))
    TN,FP,FN,TP = confusion_matrix(y_test,model[i].predict(x_test)).ravel()
    
    score= (TP+TN) / (TN+FP+FN+TP)
  
    print(cm)
    
    print('Testing Accuracy: ',score)
    disp=ConfusionMatrixDisplay(confusion_matrix=cm).plot()
    disp.ax_.set_title("Confusion Matrix- Model {}".format(i))
    
    print()

    

    
    
```
